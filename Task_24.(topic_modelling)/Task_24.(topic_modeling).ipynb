```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import gensim
```

```python
fn = 'voted-kaggle-dataset.csv'
df = pd.read_csv(fn)
```

```python
print('len of texts= {:,}'.format(len(df)))
index = 10
df.loc[index, 'Description']
```

```python
df[:5]
```

### Data regulation

```python
df_subtitle = df['Description'].astype(str).tolist()
df_subtitle[:10]
```

#### Review data

```python
print(type(df_subtitle))
print('len of documents = {:,}\n'.format(len(df_subtitle)))

df_subtitle[0]
```

#### Define custom vectorizer


```python
three_words_pattern = r"\b\w{3,}\b"

vectorizer = CountVectorizer(min_df=20,stop_words='english',token_pattern=three_words_pattern)
vectorizer.fit(df_subtitle)
```

#### Review feratures 

```python
print('len of features = {:,}\n'.format(len(vectorizer.get_feature_names_out())))
print(vectorizer.get_feature_names_out()[:40])
```

#### Vectorize data set

```python
newsgroup_data_vectorized = vectorizer.transform(df_subtitle)
print(newsgroup_data_vectorized)
```

#### Create gensim corpus


```python
corpus = gensim.matutils.Sparse2Corpus(newsgroup_data_vectorized, documents_columns=False)
[item for item in corpus][:5]
```

#### Create id2word dictionary


```python
id_map = dict((v, k) for k, v in vectorizer.vocabulary_.items())
id_map
```

#### Generate LDA model

```python
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=6, id2word=id_map, passes=25, random_state=34)
```

#### Review topics

```python
ldamodel.print_topics(num_topics=6, num_words=10)
```

## Classify the new text 

```python
new_doc = ["\n\nIt's fascinating to observe the celestial dance of celestial bodies in our solar system. The changing distances of planets from the Sun play a significant role in various phenomena. Take, for instance, the elliptical orbit of Pluto and its moon Charon. The freezing temperatures that occur are not merely a consequence of shadowing effects but are primarily influenced by their growing distance from the Sun as they follow their elliptical path.\n-- "]
print(new_doc[0])
```

```python
doc_vectorized = vectorizer.transform(new_doc)

new_doc_corpus = gensim.matutils.Sparse2Corpus(doc_vectorized, documents_columns=False)

doc_topics = ldamodel.get_document_topics(new_doc_corpus)
list(doc_topics)
```
