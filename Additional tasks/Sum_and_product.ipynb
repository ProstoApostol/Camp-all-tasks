## Task about sum and product

Задано два цілих числа:
1 < a,b <100

Один вчений ("Sum") отримує суму чисел,
іншому ("Доб") - добуток чисел. 
Обидва вчені знають, що числа 1 < a,b <100.

Визначте, які числа є добутком, виходячи з наступного діалогу: 

Прод: Я не знаю чисел;

Sum: Я знаю;

Прод: тоді я знаю числа; 

Sum: тоді я теж знаю числа.

```python
Min = 2 # Мінімальне число
Max = 100 # Максимальне число
Max_Product = 9801 # 99 * 99 = 9801
Max_Sum = 198 # 99 + 99 = 198
```

Функція для перевірки числа на просте та не просте:

```python
def is_prime(number):
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True
```

## Step 1
### Добуток: "Я не знаю чисел"

Оскільки він не зміг одразу назвати два числа, виходить, що число розкладається на множники більше ніж одним способом.

```python
def Step():

    Product = {}
    Product_Check = []
    Quantity = 0

    for q in range(Min, Max_Product, 1):
        
        for i in range(Min, Max, 1):
           for w in range(i + 1, Max, 1):
                if i * w == q:
                        Product_Check.append((i,w))
                        Quantity = Quantity + 1

        if Quantity >= 2:
            Product[q] = Product_Check

        Quantity = 0
        Product_Check = []

    return Product

Product = Step()
for key, value in Product.items():
    if key < 60:                  # Для зменшення кількості виведених даних на екран.
        print(f"{key}: {value}")
        
print("Len: ",len(Product))
```

## Step 2
### Сума: "Я знаю це"

Тут потрібно, щоб за будь-якого розбиття на доданки, більші за одиницю, добуток доданків має розкладатися на множники більш ніж одним способом, та сума числа не могла б бути представлена у вигляді двох простих чисел.

Виявляється, що якщо будь-яке просте число, яке є більшим за половину максимально допустимого числа в задачі (в нашому випадку це 100), є одним із множників, то добуток буде унікальним числом. Саме тому всі суми, більші за 53, відкидаються, оскільки число 53 є простим і є більшим ніж половина числа 100.

```python
def Step2():

    Sum = {}
    Sum_Check = []
    Quantity = 0
    Check = 0

    for q in range(Min, Max_Sum, 1):
        
        if q > 53:
            break
        
        for i in range(Min, Max, 1):
            for w in range(i + 1, Max, 1):

                if i + w == q:

                    Sum_Check.append((i, w))
                    Quantity += 1

                elif i + w > q:
                    break

        if Quantity >= 2:

            for a in Sum_Check:

                for s in a:
                    if is_prime(s):
                        Check += 1

                if Check >= 2:
                    break

                Check = 0

            if Check < 2:
                Sum[q] = Sum_Check

        Check = 0
        Quantity = 0
        Sum_Check = []

    return Sum

Sum = Step2()

number = (5,10)        
for key, value in Sum.items():
        print(f"{key}: {value}")
print("Len: ",len(Sum))
```

## Step 3
### Добуток: "Я знаю числа"

Тому що з усіх відомих йому розкладень рівно одне дає суму, яка задовольняє вимогу з попереднього абзацу.

Наприклад, якщо султан загадав 2 і 9, то з двох розкладів на множники тільки одне, 2*9, дає суму, що має потрібну властивість (2+9=11), а друге, 3*6, - ні (тому що 3+6=9).

```python
def Step3(Product, Sum):

    Product_and_Sum = {}
    Product_and_Sum_Check = {}
    Check = 0

    for key_P, value_P in Product.items():

        for i in value_P:

            for key_W, value_S in Sum.items():
                if i in value_S:
                    Product_and_Sum_Check[key_W] = i # Для зручності в наступній операції нехай ключем буде сума чисел
                    Check += 1

        if Check == 1:
            for key, value in Product_and_Sum_Check.items():
                if key in Product_and_Sum:
       
                    Product_and_Sum[key].append(value)
                else:
        
                    Product_and_Sum[key] = [value]

        Product_and_Sum_Check = {}
        Check = 0

    return Product_and_Sum

Product_and_Sum = Step3(Product, Sum)
for key, value in Product_and_Sum.items():
        print(f"{key}: {value}")
print("Len: ",len(Product_and_Sum))
```

## Step 4
### Сума: "Я теж знаю числа"

Якби йому було відомо, що сума дорівнює 11, він міг би відгадати числа? Очевидно, що ні, оскільки в нього не було б жодних підстав вибрати між варіантами 2+9, 3+8 і 4+7.

Отже, виходить, що з усіх розбивок суми на доданки рівно одна повинна задовольняти умову. Наприклад, це може бути сума 17 з розкладанням 4+13=17.

```python
def Step4(Product_and_Sum):
    
    Result = {}
    
    for key,value in Product_and_Sum.items():
        if len(Product_and_Sum[key]) == 1:
            Result[key] = value

    return Result

Result = Step4(Product_and_Sum)
for key, value in Result.items():
    print(f"{key}: {value}")
```
